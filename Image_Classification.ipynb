```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

## Load the image using cv2

```python
im=cv2.imread("girl2.jpg")
```

```python
im.shape
```

## Converting the RGB image a grayscale image

```python
im=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
im.shape
```

## Define vertical and horizontal filter kernels

```python
v_kernel=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
h_kernel=np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
```

## Do the filtering

```python
v_image=cv2.filter2D(im,-1,v_kernel)
h_image=cv2.filter2D(im,-1,h_kernel)
```

## Display the result

```python
plt.imshow(im,cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image,cmap="gray")
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image,cmap="gray")
plt.title('Horizontal Image')
```
