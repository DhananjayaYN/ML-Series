```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Scalered_data = scaler.fit_transform(digits.data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)

import time 

start = time.time()
new_digits = pca.fit_transform(Scalered_data)
end = time.time()

new_digits.shape
```

```python
end-start
```
